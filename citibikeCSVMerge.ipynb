```python
##Jersey city csv merge
import pandas as pd 
```

```python
citibikes_data  = ["data/202208.csv","data/202209.csv","data/202210.csv","data/202211.csv","data/202212.csv"]
```

```python
citibikes_list = []

for data in citibikes_data:

    citibikes = pd.read_csv(data)

    citibikes_list.append(citibikes)

    citibikesDF = pd.concat(citibikes_list)

citibikesDF
```

```python
citibikesDF.dropna(inplace=True)
citibikesDF
```

```python
citibikesDF.to_csv("data/citibikes_2022_merge.csv", index=False)
```
